```python
from urllib.request import urlretrieve
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
urlretrieve(medical_charges_url)
medical_charges_df = pd.read_csv('medical.csv')
```

```python
non_smokers_df = medical_charges_df[medical_charges_df.smoker == 'no']
```

```python
plt.scatter(non_smokers_df.age, non_smokers_df.charges, s=8)
plt.xlabel("Age")
plt.ylabel("Charges");
```

```python
import numpy as np
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))
```

```python
def estimate_line(x, a, b):
    return a*x+b
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
 inputs = non_smokers_df[['age']] #x
outputs = non_smokers_df.charges #y
model.fit(inputs, outputs);
```

```python
a = model.coef_
b = model.intercept_
```

```python
ages = non_smokers_df.age
target = non_smokers_df.charges
```

```python
estimated_charges = estimate_line(ages, a, b)
```

```python
plt.title(f"Annuel Medical Charges and RMSE = {rmse(target,       estimated_charges)}")
plt.plot(ages, estimated_charges, 'r', alpha=0.9);
plt.scatter(ages, target, s=8,alpha=0.8);
plt.xlabel('Age');
plt.ylabel('Charges')
plt.legend(['Estimate', 'Actual']);
```
